# Road Accidents India : 2017 - 2020

The dataset is taken from the GOI site for Road and Transport Ministry for public safety.It reflects the number of accidents , their nature and the fatalities . It also looks into the numbers , shares and unit wise distribution of the calamity.The time span is from 2017 - 2020

Dataset link :https://www.kaggle.com/datasets/zsinghrahulk/road-accidents-india-2017-2020

### Objective:

- Find Which State has highest Number of Road accident and plot the Graph of highest 20 State For All 4 Years.

- Find and Plot 5 state with least number of deathFor All 4 Years .

- Find Which State has highest Total Number of accident on lakh population For All 4 Years.

- Which State has highest Number of People Death in Road Accident For All 4 Years.

- Which State has highest Number of Injuiruies for All 4 Years.


## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Downloading Dataset

```python
df=pd.read_csv("India_Injury_Road_Accident_Fatality_2017-2020.csv")
df.head()
```

## Data Preparation and Cleaning

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(axis=0,how='any',inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.columns=[i.replace(' ','_')for i in df.columns]
df.columns
```

## Exploratory Data Analysis

```python
df.sort_values(by='Road_Accidents__during_2018',ascending=False)
```

### Find Which State has highest Number of Road accident in Year 2018 and plot the Graph of highest 20 State.

```python
highest_acc_2018=df.sort_values('Road_Accidents__during_2018',ascending=False)[1:21]
highest_acc_2018
```

```python
plt.figure(figsize=(15,9))
sns.barplot(data=highest_acc_2018,x='State',y='Road_Accidents__during_2018')
plt.title("Highest Accident in 2018 State Wise",fontsize=14,pad=14)
plt.xlabel('State',fontsize=12)
plt.ylabel('Accidents During 2018',fontsize=12)

plt.tick_params(axis='x',which='major',labelsize=12,rotation=90)
plt.tick_params(axis='y',which='major',labelsize=12)
```

- Tamil Nadu has highest Number of Road accidents in year 2018. The possible reasons are the large poplulation that are residing in the state

### Find and Plot 5 state with least number of death in 2018

```python
low_acc_2018=df.sort_values(by='Road_Accidents__during_2018')[:5]
low_acc_2018
```

```python
plt.figure(figsize=(15,9))
sns.barplot(data=low_acc_2018,x='State',y='Road_Accidents__during_2018')
plt.title("Lowest Accident in 2018 State Wise",fontsize=14,pad=14)
plt.xlabel('State',fontsize=12)
plt.ylabel('Accidents During 2018',fontsize=12)

plt.tick_params(axis='x',which='major',labelsize=12)
plt.tick_params(axis='y',which='major',labelsize=12)
```

- The reasons for the lowest accident in these states are less population and presence of less vehicles.
